```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from skimpy import skim
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
skim(titanic)
```

```python
titanic.drop('deck',axis=1,inplace=True)
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0],inplace=True)
titanic['embark_town'].fillna(titanic['embark_town'].mode()[0],inplace=True)
```

```python
titanic.sample(20)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in titanic.columns:
    if titanic[col].dtype == 'object' or titanic[col].dtype.name == 'category':
        titanic[col]=LabelEncoder().fit_transform(titanic[col])
titanic.head()
```

```python
sns.countplot(x='class', data=titanic)
plt.title('Passenger Count by Class')
plt.show()
```

```python
# Histogram: Distribution of ages
sns.histplot(titanic['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Pie chart: Proportion of males and females
gender_counts = titanic['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
# Scatter plot: Fare vs. Age
sns.scatterplot(x='age', y='fare', data=titanic)
plt.title('Scatter Plot of Fare vs. Age')
plt.show()
```

```python
X=titanic.drop('survived',axis=1)
y=titanic['survived']
```

```python
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
clf.fit(x_train,y_train)
```

```python
prediction=clf.predict(x_test)
prediction
```

```python
from sklearn.metrics import accuracy_score,recall_score, precision_score, f1_score,confusion_matrix, classification_report
```

```python
print('Accuracy Score:',accuracy_score(y_test,prediction))
print('Recall Score: ',recall_score(y_test,prediction))
print('Precision Score: ',precision_score(y_test,prediction))
print('F1 Score: ',f1_score(y_test,prediction))
print('Confusion Matrix: \n',confusion_matrix(y_test,prediction))
print('Classification Report: \n',classification_report(y_test,prediction))
```
